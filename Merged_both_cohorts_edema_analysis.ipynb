```python
def core_extractor(var):
    if isinstance(var,tuple):
        return var[0]
    else:
        return var
    
def penumbra_extractor(var):
    if isinstance(var,tuple):
        if len(var)>1:
            return var[1]
        else:
            return 0
    else:
        return var
```

```python
#New columns for *automated* core and penumbra volumes

both_init_cohorts["RAPID Core (mL)"] = both_init_cohorts["volumes"].apply(core_extractor)
both_init_cohorts["RAPID Penumbra (mL)"] = both_init_cohorts["volumes"].apply(penumbra_extractor)
#both_init_cohorts = both_init_cohorts[both_init_cohorts[['RAPID Core (mL)', 'RAPID Penumbra (mL)']].notnull().all(1)]
both_init_cohorts
```

```python
def lit(var):
    if isinstance(var,str):
        return ast.literal_eval(var)
    else:
        return var

# for no in both_init_cohorts.index.values:
#     print(str(lit(both_init_cohorts.pred_segments.loc[no]))+" "+str(no))
```

```python
both_init_cohorts["pred_vessels"] = both_init_cohorts.pred_vessels.apply(lit)
both_init_cohorts["pred_segments"] = both_init_cohorts.pred_segments.apply(lit)
both_init_cohorts[both_init_cohorts["pred_occlusion"]==1]
```

```python
#USING NLP PREDICTIONS
#0=no occlusion
#1=proximal occlusion
#2=distal occlusion
def check_proximal_distal(occlusion_status, vessels, segments):
    is_occlusion = occlusion_status
    if is_occlusion > 0:
        if any("ICA" in s for s in vessels):
            return 1
        if segments == []:
            return 0
        if any("MCA" in s for s in vessels):
            if any("M1" in s for s in segments):
                return 1
            return 2
    return 0
```

```python
#Check prox vs dist of NLP
both_init_cohorts["proximal_occlusion"] = both_init_cohorts.apply(lambda x: check_proximal_distal(x.pred_occlusion,x.pred_vessels,x.pred_segments), axis=1)
both_init_cohorts[both_init_cohorts["proximal_occlusion"]>0]
```

```python
from scipy.stats import norm

ds1 = e[e["proximal_occlusion"]==1]["midline shift (mm)"]
ds2 = e[e["proximal_occlusion"]==2]["midline shift (mm)"]

ct1 = ds1.count()  #items in dataset 1
ct2 = ds2.count()  #items in dataset 2
alpha = 0.05       #95% confidence interval
N = norm.ppf(1 - alpha/2) # percent point function - inverse of cdf

# The confidence interval for the difference between the two population
# medians is derived through these nxm differences.
diffs = sorted([i-j for i in ds1 for j in ds2])

# For an approximate 100(1-a)% confidence interval first calculate K:
k = int(round(ct1*ct2/2 - (N * (ct1*ct2*(ct1+ct2+1)/12)**0.5)))

# The Kth smallest to the Kth largest of the n x m differences 
# ct1 and ct2 should be > ~20
CI = (diffs[k], diffs[len(diffs)-k])
CI
```

```python
import numpy as np
import scipy.stats as stats

def non_param_unpaired_CI(sample1, sample2, conf):
  n1 = len(sample1)  
  n2 = len(sample2)  
  alpha = 1-conf      
  N = stats.norm.ppf(1 - alpha/2) 

  # The confidence interval for the difference between the two population
  # medians is derived through the n x m differences.
  diffs = sorted([i-j for i in sample1 for j in sample2])
  
  # the Kth smallest to the Kth largest of the n x m differences then determine 
  # the confidence interval, where K is:
  k = np.math.ceil(n1*n2/2 - (N * (n1*n2*(n1+n2+1)/12)**0.5))
  
  CI = (round(diffs[k-1],3), round(diffs[len(diffs)-k],3))
  return CI

non_param_unpaired_CI(ds1,ds2, .05)
```
